Implementing basic perceptron model for binary classification

```python
import numpy as np
import matplotlib.pyplot as plt
```

Generate a simple binary classification dataset

Class 0: centered at (2,2)

Class 1: centered at (4,4)

```python
np.random.seed(0)
class0 = np.random.randn(50,2) + [2,2]
class1 = np.random.randn(50,2) + [4,4]
```

```python
X = np.vstack((class0, class1)) # features
y = np.hstack((np.zeros(50), np.ones(50))) # labels: 0 and 1
```

```python
# Add bias term (x0 = 1)
X_bias = np.hstack((np.ones((X.shape[0], 1)),X))
```

```python
# initialize the weights, learning rate and epochs
weights = np.zeros(X_bias.shape[1])
learning_rate = 0.1
epochs = 10
```

# Activation function

```python
# Step activation function
def step_function(x):
  return np.where(x >= 0, 1, 0)
```

Training the perceptron

```python
for epoch in range(epochs):
  for xi, target in zip(X_bias, y):
    output = step_function(np.dot(xi, weights))
    error = target - output
    weights += learning_rate * error * xi
print('Final weights: ', weights)
```

Plotting the decision boundary

```python
plt.figure(figsize=(8,6))
plt.scatter(class0[:,0], class0[:,1], color='red', label='Class 0')
plt.scatter(class1[:,0], class1[:,1], color='blue', label='Class 1')

# Decision boundary: w0 + w1*x + w2*y = 0 -> y = -(w0 + w1*x)/w2
x_vals = np.linspace(0,6,100)
y_vals = -(weights[0] + weights[1]*x_vals)/weights[2]
plt.plot(x_vals, y_vals, color='black', label='Decision Boundary')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('perceptron Binary Classification')
plt.legend()
plt.grid(True)
plt.show()
```
