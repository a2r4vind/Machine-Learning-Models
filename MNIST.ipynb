```python
# Since the above output is of string type so casting it into integer type
import numpy as np
y = y.astype(np.uint8)
y[0]
```

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
y_train_5 = (y_train == 5)
y_test_5 = (y_test == 5)
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_5)
```

```python
sgd_clf.predict([some_digit])
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone
```

```python
skfolds = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)

for train_index, test_index in skfolds.split(X_train, y_train_5):
  clone_clf = clone(sgd_clf)
  X_train_folds = X_train.iloc[train_index]
  y_train_folds = y_train_5.iloc[train_index]
  X_test_fold = X_train.iloc[test_index]
  y_test_fold = y_train_5.iloc[test_index]

  clone_clf.fit(X_train_folds, y_train_folds)
  y_pred = clone_clf.predict(X_test_fold)
  n_correct = sum(y_pred == y_test_fold)
  print(n_correct / len(y_pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring='accuracy')
```

```python
from sklearn.base import BaseEstimator
```

```python
class Never5Classifier(BaseEstimator):

  def fit(self, X, y=None):
    return self

  def predict(self, X):
    return np.zeros((len(X), 1), dtype=bool)
```

For classifiers **Accuracy** is generally not the preferred performance measure especially when we are dealing with the **skewed dataset**.

```python
never_5_clf = Never5Classifier()
cross_val_score(never_5_clf, X_train, y_train_5, cv=3, scoring='accuracy')
```

FP - False Positives

Recall = Sensitivity = True Positve rate

# F1 score

To compare two classifiers F1 score is one of the best measure which combine Precision and Recall by calculating *harmonic mean* of both.

Regular mean treats all values equally but harmonic mean gives more weight to low values

==> High F1 score when high Precion and Recall

```python
from sklearn.metrics import f1_score
```

```python
f1_score(y_train_5, y_train_pred)
```

```python
y_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3, method='decision_function')
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
precision, recall, thresholds = precision_recall_curve(y_train_5, y_scores)
```

```python
def plot_precision_recall_vs_threshold(precision, recall, thresholds):
  plt.plot(thresholds, precision[:-1], "b--", label="Precision")
  plt.plot(thresholds, recall[:-1], "g-", label="Recall")
  plt.xlabel("Thresholds")
  plt.legend(loc='upper left')
  plt.grid(True)
  plt.ylim([0,1])
  plt.show()

plot_precision_recall_vs_threshold(precision, recall, thresholds)
```

```python
threshold_90_precision = thresholds[np.argmax(precision >= 0.90)]
threshold_90_precision
```

```python
y_train_pred_90 = (y_scores >= threshold_90_precision)
```

```python
precision_score(y_train_5, y_train_pred_90)
```

```python
recall_score(y_train_5, y_train_pred_90)
```

```python
from sklearn.metrics import roc_curve
```

```python
fpr, tpr, thresholds = roc_curve(y_train_5, y_scores)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
  plt.plot(fpr, tpr, linewidth=2, label=label)
  plt.plot([0,1], [0,1], 'k--')
  plt.axis([0,1, 0, 1])
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.grid(True)
  plt.show()
```

```python
plot_roc_curve(fpr, tpr)
```
