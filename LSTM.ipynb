Using Long Short Term Memory (LSTM) cells for sequence predictions using TensorFlow.

Necessary imports

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

Generate synthetic data

```python
np.random.seed(42)
sequence_length = 10
num_samples = 1000
input_dim = 1
```

```python
X = np.random.random((num_samples, sequence_length, input_dim))
y = np.sum(X, axis=1)
```

Split the data into trainging and testing sets

```python
split_ratio = 0.8
split_index = int(split_ratio * num_samples )
X_train, X_test = X[:split_index], X[split_index : ]
y_train, y_test = y[:split_index], y[split_index : ]
```

Build the LSTM model

```python
model = Sequential()
model.add(LSTM(10, input_shape=(sequence_length, input_dim), activation='relu'))
model.add(Dense(1))
```

Compile the model

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

Train the model

```python
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```

Evaluate the model on the test data

```python
test_loss = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}")
```

Make predictions using the trained model

```python
predictions = model.predict(X_test)
```

print few predictions and corresponding actual values

```python
for i in range(5):
  print(f"Predicted Value : {predictions[i][0]}, Actual Value: {y_test[i]}")
```
