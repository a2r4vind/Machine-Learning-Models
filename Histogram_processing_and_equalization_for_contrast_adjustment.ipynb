```python
# if image not found, create synthetic image
# if img is None:
img = np.zeros((256,256), dtype=np.uint8)
cv2.rectangle(img, (50,50), (200,200), 120, -1)
cv2.circle(img, (128, 128), 60, 200, -1)
```

```python
# Contrast Stretching function
def contrast_stretching(image):
  min_val = np.min(image)
  max_val = np.max(image)
  stretched = ((image - min_val) * (255/(max_val - min_val))).astype(np.uint8)
  return stretched
```

```python
# 1. Contrast adjustment (linear contrast stretch)
contrast_img = contrast_stretching(img)
```

```python
# 2. Histogram equalization
equalized_img = cv2.equalizeHist(img)
```

```python
# plot original, contrast adjusted and equalized images with histograms
plt.figure(figsize=(15,10))
```

```python
# Original Image and histogram
plt.subplot(3,2,1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(3,2,2)
plt.title('Histogram of Original Image')
plt.hist(img.ravel(), bins=256, range=[0,256])
plt.xlim([0,256])
```

```python
# Contrast stretched image and histogram
plt.subplot(3,2,3)
plt.title('Contrast Stretched Image')
plt.imshow(contrast_img, cmap='gray')
plt.axis('off')

plt.subplot(3,2,4)
plt.title('Histogram of Contrast Stretched Image')
plt.hist(contrast_img.ravel(), bins=256, range=[0,256])
plt.xlim([0,256])
```

```python
# Histogram equalized image and histogram

plt.subplot(3,2,5)
plt.title('Histogram Equalized Image')
plt.imshow(equalized_img, cmap='gray')
plt.axis('off')

plt.subplot(3,2,6)
plt.title('Histogram of Histogram Equalized Image')
plt.hist(equalized_img.ravel(), bins=256, range=[0,256])
plt.xlim([0,256])
```
